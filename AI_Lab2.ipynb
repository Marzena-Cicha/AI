```python
import tensorflow as tf
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
print(tf.__version__)
```

```python
training_data = np.load('TF_dataset.npz')
input_size = 2
output_size = 1
```

```python
model = tf.keras.Sequential([tf.keras.layers.Dense(output_size)])

model.compile(optimizer='sgd', loss='mean_squared_error')
print("DostД™pne klucze w training_data:", training_data.files)
model.fit(training_data['input'], training_data['targets'], epochs=100, verbose=1)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(5, activation="sigmoid"),
    tf.keras.layers.Dense(output_size)
])

model.compile(optimizer='sgd', loss='mean_squared_error')
print("DostД™pne klucze w training_data:", training_data.files)
model.fit(training_data['input'], training_data['targets'], epochs=100, verbose=1)
```

```python
import matplotlib.pyplot as plt
plt.plot(np.squeeze(model.predict_on_batch(training_data['input'])),
np.squeeze(training_data['targets']), color='magenta')
plt.xlabel('outputs')
plt.ylabel('targets')
plt.show()
```
